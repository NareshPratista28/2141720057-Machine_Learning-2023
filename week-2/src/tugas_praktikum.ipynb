> ### 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.



- Import library dan load data
- memisahkan variabel dengan *columns_to_exclude*

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv('wbc.csv')
df.head(100)
```

```python
columns_to_exclude = ['Unnamed: 32', 'id']
filtered = df.drop(columns=columns_to_exclude)
filtered.head()
```

> ### 2. Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

df.head()
```

> ### 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

- melakukan standarisasi pada kolom-kolom numerik dalam dataset *filtered*

```python
sc = StandardScaler()
nc = filtered.columns.difference(['diagnosis'])
filtered[nc] = sc.fit_transform(filtered[nc])
filtered.head(10)
```

> ### 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

- x untuk menghapus kolom diagnosis variabel fitur
- y sebagai target dari variabelnya
- stratisfied split data dengan ratio 80:20
- menampilkan banyaknya data latih dan test

```python
x = filtered.drop('diagnosis', axis = 1)
y = filtered['diagnosis']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=42)

print("Banyaknya data latih atau x_train  :", len(x_train))
print("Banyaknya data test atau x_test    :", len(x_test))
```
